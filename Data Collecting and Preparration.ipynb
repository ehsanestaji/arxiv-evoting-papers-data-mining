```python
# Final Collector Function 
from tqdm import tqdm

def Lexicon_List_Wrapper():
    f = open("lexicon.txt", "r")
    lex_list = []
    for lines in f.readlines():
        if len(lines) > 0:
            title = re.findall(r'\"(.+?)\"', lines)
            lex_list.append(title[0])
    return lex_list



# Defining GRabber Function
def PaperGrabber(keyword): 
    headers = {'User-agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36 Edge/18.19582"}
    params = { "q": keyword, "hl": "en"}
    html = requests.get('https://scholar.google.com/scholar', headers=headers, params=params).text
    soup = BeautifulSoup(html, 'lxml')
    j= 80 
    #URL = "https://scholar.google.com/scholar?start="+str(j)+"&q=receipt+freeness&hl=en"
    time.sleep((30-5)*np.random.random()+10) #from 5 to 30 seconds
    #page = requests.get(URL)
    #soup = BeautifulSoup(page.content, 'html.parser')
    data = [] 
    for result in soup.select('.gs_ri'):
        title = result.select_one('.gs_rt').text
        title_link = result.select_one('.gs_rt a')['href']
        publication_info = result.select_one('.gs_a').text
        snippet = result.select_one('.gs_rs').text
        cited_by = result.select_one('#gs_res_ccl_mid .gs_nph+ a').text
        cited_by = re.findall('\d+', cited_by)[0]
        related_articles = result.select_one('a:nth-child(4)')['href']
        data.append([title , title_link, publication_info, snippet, cited_by,related_articles])
    df = pd.DataFrame(data, columns = ['title', 'title_link','publication_info','snippet','cited_by','related_articles']) 
    return df



def FinalCollector(_from,_to):
    dff=pd.DataFrame()
    for word in tqdm(Lexicon_List_Wrapper()[_from:_to]):
        dff = dff.append(PaperGrabber(word), ignore_index= True)
    filename = "data_from_"+str(_from)+"_to_"+str(_to)+".csv"
    dff.to_csv(filename)
    return dff
```

```python
# Merging all the csv files in the one dataframe
import glob 

def FileMarger():
    files = os.path.join("Collected CSV Files/data_from*.csv")
    files = glob.glob(files)
    df = pd.concat(map(pd.read_csv, files), ignore_index=True)
    df.drop('Unnamed: 0', axis = 1, inplace= True)
    print("Mission Completed  ! ")
    return df
```
